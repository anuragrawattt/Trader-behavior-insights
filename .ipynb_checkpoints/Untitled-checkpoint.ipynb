```python
import pandas as pd
import numpy as np
f = pd.read_csv("fear_greed_index.csv")
df = pd.read_csv("historical_data.csv")
print("Number of columns of fear greed:", f.shape[1])
print("Number of rows of fear greed:", f.shape[0])
print("Number of columns of traders:", trades.shape[1])
print("Number of rows of traders:", trades.shape[0])
print("FGI Missing Values:\n", f.isnull().sum())
print("Trades Missing Values:\n", trades.isnull().sum())
print("Fear and greed Duplicates:", f.duplicated().sum())
print("Trades Duplicates:", trades.duplicated().sum())


# Fear Greed timestamp convert

f['timestamp'] = pd.to_datetime(f['timestamp'], unit='s')
#Convert date (string DD-MM-YYYY â†’ datetime)
f['date'] = pd.to_datetime(f['date'], format='%d-%m-%Y')
f_daily = (f.groupby(f['date']).mean(numeric_only=True).reset_index())
f.to_csv("fear_greedupdated.csv", index=False)


#for Traders historical data

df['Timestamp IST'] = pd.to_datetime(df['Timestamp IST'], format='%d-%m-%Y %H:%M')
df['Date'] = df['Timestamp IST'].dt.date


#The key metrics

df['Win Rate'] = df['Closed PnL'] > 0
df['Start Position'] = pd.to_numeric(df['Start Position'], errors='coerce')

#because some of the values have start position as 0 so division by zero will be 'inf' replacing it with nan
df['Leverage'] = df['Size USD'] / df['Start Position']
df['Leverage'] = df['Leverage'].replace([np.inf, -np.inf], np.nan)

summary = df.groupby(['Date','Account']).agg({
    'Closed PnL':'sum',          # Daily PnL
    'Win Rate':'mean',                # Win rate
    'Size USD':'mean',
    'Size Tokens':'mean',        # Avg trade size (Tokens)
    'Fee':'sum',                 # Total fees
    'Side':'count',      # Number of trades
    'Leverage':'mean',           # Avg leverage
}).reset_index()
summary['LongTrades'] = df.groupby(['Date','Account'])['Side'].apply(lambda x: (x=='BUY').sum()).values
summary['ShortTrades'] = df.groupby(['Date','Account'])['Side'].apply(lambda x: (x=='SELL').sum()).values
summary['LongShortRatio'] = summary['LongTrades'] / (summary['ShortTrades']+1)
df = df.drop(columns=['Win Rate','Leverage'], errors='ignore')
df = df.merge(summary, on=['Date','Account'], how='left', suffixes=('', '_daily'))
df = df.rename(columns={
    'Win Rate_x':'Win Rate (raw)',
    'Win Rate_y':'Win Rate (daily)',
    'Leverage_x':'Leverage (raw)',
    'Leverage_y':'Leverage (daily)',
    'Closed PnL_y':'Closed PnL (daily)'
})
df.to_csv("traderupdated.csv", index=False)

# df.to_csv("traderupdated.csv", index=False)
# summary.to_csv("traderupdated.csv", index=False)
# import matplotlib.pyplot as plt

# plt.figure()
# merged.groupby('value')['PnL'].mean().plot(kind='bar')
# plt.title("Average PnL vs Fear-Greed Index")
# plt.ylabel("PnL")
# plt.show()
```
